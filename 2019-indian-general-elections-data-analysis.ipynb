**Question 1: Which States/UTs and Constituencies had the highest and the lowest Voter Turnout?**

```python
# Extract the highest and lowest voter turnout constituencies
top_5_highest_turnout = df.nlargest(5, 'OVER TOTAL ELECTORS IN CONSTITUENCY')
top_5_lowest_turnout = df.nsmallest(5, 'OVER TOTAL ELECTORS IN CONSTITUENCY')

# Visualization for Top 5 Highest Voter Turnout
plt.figure(figsize=(14, 7))
sns.barplot(x='OVER TOTAL ELECTORS IN CONSTITUENCY',y='CONSTITUENCY',hue='STATE',
            data=top_5_highest_turnout,
            dodge=False
)
plt.title('Top 5 Constituencies with Highest Voter Turnout')
plt.xlabel('Voter Turnout (%)')
plt.ylabel('Constituency')
plt.legend(title='State')
plt.tight_layout()
plt.show()

# Visualization for Top 5 Lowest Voter Turnout
plt.figure(figsize=(14, 7))
sns.barplot(x='OVER TOTAL ELECTORS IN CONSTITUENCY',y='CONSTITUENCY',hue='STATE',
    data=top_5_lowest_turnout,
    dodge=False
)
plt.title('Top 5 Constituencies with Lowest Voter Turnout')
plt.xlabel('Voter Turnout (%)')
plt.ylabel('Constituency')
plt.legend(title='State')
plt.tight_layout()
plt.show()
```

**Two bar plots showing the constituencies with the highest and lowest voter turnout percentages. These plots highlight the regions with the most and least engaged electorates, suggesting areas of high political activity, respectively.**

**Question 2: How many elected MPs have a criminal record?**

```python
# Convert 'CRIMINAL CASES' to numeric, setting errors='coerce' to handle non-numeric values
df['CRIMINAL CASES'] = pd.to_numeric(df['CRIMINAL CASES'], errors='coerce').fillna(0).astype(int)

# Criminal Records of Elected MPs
elected_mps = df[df['WINNER'] == 1]
criminal_record_mps = elected_mps[elected_mps['CRIMINAL CASES'] > 0]

# Select top 10 MPs with the highest number of criminal cases
top_10_criminal_record_mps = criminal_record_mps.nlargest(10, 'CRIMINAL CASES')

# Visualization for top 10 MPs with criminal records
plt.figure(figsize=(14, 7))
sns.barplot(x='CRIMINAL CASES',y='NAME',hue='PARTY',data=top_10_criminal_record_mps,dodge=False)
plt.title('Top 10 MPs with Criminal Records')
plt.xlabel('Number of Criminal Cases')
plt.ylabel('MP Name')
plt.legend(title='Party')
plt.tight_layout()
plt.show()
```

**A bar plot showing the number of elected MPs with criminal records, categorized by the number of criminal cases. This visualization exposes the prevalence of criminal records among elected MPs, raising concerns about the integrity of some elected officials.**

**Question 2: criminal record of MPs and their party-wise distribution?**

```python
# Party-wise distribution
party_wise_criminal_records = criminal_record_mps['PARTY'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=party_wise_criminal_records.index, y=party_wise_criminal_records.values)
plt.title('Party-wise Distribution of MPs with Criminal Records')
plt.xlabel('Party')
plt.xticks(rotation=45)
plt.ylabel('Number of MPs with Criminal Records')
plt.show()
```

**A bar plot showing the distribution of MPs with criminal records by party. It provides a breakdown of which parties have more MPs with criminal backgrounds, potentially impacting the party's image.**

**Question 3: What is the distribution of assets, liabilities, and net worth of the elected MPs?**

```python
# Ensure the columns are numeric, converting and filling NaN values with 0
df['ASSETS'] = pd.to_numeric(df['ASSETS'], errors='coerce').fillna(0)
df['LIABILITIES'] = pd.to_numeric(df['LIABILITIES'], errors='coerce').fillna(0)

# Calculate 'NET WORTH'
df['NET WORTH'] = df['ASSETS'] - df['LIABILITIES']

# Filter the DataFrame to include only elected MPs
elected_mps = df[df['WINNER'] == 1]

# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plotting assets distribution
axes[0].hist(elected_mps['ASSETS'], bins=30, edgecolor='black', alpha=0.7)
axes[0].set_title('Distribution of Assets')
axes[0].set_xlabel('Assets')
axes[0].set_ylabel('Frequency')

# Plotting liabilities distribution
axes[1].hist(elected_mps['LIABILITIES'], bins=20, edgecolor='black', alpha=0.7)
axes[1].set_title('Distribution of Liabilities')
axes[1].set_xlabel('Liabilities')
axes[1].set_ylabel('Frequency')

# Plotting net worth distribution
axes[2].hist(elected_mps['NET WORTH'], bins=20, edgecolor='black', alpha=0.7)
axes[2].set_title('Distribution of Net Worth')
axes[2].set_xlabel('Net Worth')
axes[2].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

**Three histograms showing the distribution of assets, liabilities, and net worth among elected MPs. These plots reveal the financial status of elected MPs.**

**Question 4: Which 10 constituencies saw the Most One-sided and the Closest victories respectively?**

```python
# Most One-sided and Closest Victories
df['MARGIN'] = df['GENERAL VOTES'] - df['POSTAL VOTES']
most_onesided_victories = df.nlargest(10, 'MARGIN')
closest_victories = df.nsmallest(10, 'MARGIN')

# Visualization for Most One-sided Victories
plt.figure(figsize=(14, 7))
sns.barplot(x='MARGIN', y='CONSTITUENCY',data=most_onesided_victories)
plt.title('Top 10 Most One-sided Victories')
plt.xlabel('Margin of Victory')
plt.ylabel('Constituency')
plt.tight_layout()
plt.show()

# Visualization for Closest Victories
plt.figure(figsize=(14, 7))
sns.barplot(x='MARGIN',y='CONSTITUENCY',data=closest_victories)
plt.title('Top 10 Closest Victories')
plt.xlabel('Margin of Victory')
plt.ylabel('Constituency')
plt.tight_layout()
plt.show()
```

**Two bar plots showing constituencies with the most significant and smallest margins of victory. These visualizations identify the areas where elections were highly competitive or overwhelmingly one-sided, providing insight into electoral dynamics.**

**Question 5: What is the region-wise distribution of seats across the country?**

```python
# Region-wise Distribution of Seats
regions = {
    'NORTH INDIA': ['Jammu & Kashmir', 'Himachal Pradesh', 'Punjab', 'Haryana', 'Uttarakhand', 'NCT OF Delhi', 'Chandigarh'],
    'CENTRAL INDIA': ['Uttar Pradesh', 'Madhya Pradesh', 'Chhattisgarh'],
    'WEST INDIA': ["Gujarat", "Rajasthan", "Maharashtra", "Goa", "Dadra & Nagar Haveli", "Daman & Diu"],
    'SOUTH INDIA': ['Karnataka', 'Andhra Pradesh', 'Telangana', 'Kerala', 'Tamil Nadu', 'Puducherry', 'Lakshadweep'],
    'EAST INDIA': ["Bihar", "Jharkhand", "West Bengal", "Odisha", "Andaman & Nicobar Islands"],
    'NORTH-EAST INDIA': ['Assam', 'Arunachal Pradesh', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Sikkim', 'Tripura']
}

def get_region(state):
    for region, states in regions.items():
        if state in states:
            return region
    return 'UNKNOWN'

df['REGION'] = df['STATE'].apply(get_region)

region_wise_distribution = df['REGION'].value_counts()

plt.figure(figsize=(12, 7))
sns.barplot(x=region_wise_distribution.index, y=region_wise_distribution.values)
plt.title('Region-wise Distribution of Seats')
plt.xlabel('Region')
plt.ylabel('Number of Seats')
plt.xticks(rotation=45)
plt.show()
```

**A bar plot showing the distribution of seats across different regions of India.It illustrates the geographical spread of electoral seats, showing which regions have more representation in the election results.**
